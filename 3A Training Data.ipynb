## Simple charts of each feature

```python
cols_not_dealt_with = 0
target_df = df.SK_ID_CURR
for column in df:
    
    ################################################################
    ## print some stats on the column ##############################
    ################################################################
    unique_values = df[column].nunique()
    null_values = df[column].isnull().mean()*100
    print(column, " - unique = ", unique_values, ". null values = ", null_values, "%. Type = ", df[column].dtype)
    if unique_values<10:
        print(df[column].unique())
    try:
        print(data_dictionary.loc[(data_dictionary.Table=='application_{train|test}.csv') & (data_dictionary.Row==column)]['Description'].values[0])
    except:
        print("description not found")
    
    ################################################################
    ## show the chart and decide on proposed feature encoding ######
    ################################################################
    if df[column].dtype=='object':
        sns.countplot(data=df, x=column)
        plt.xticks(rotation=45)
        if unique_values == 2 and 'Y' in df[column].unique() and 'N' in df[column].unique():
            proposed_encoding = 'Binary'
        else:
            proposed_encoding = 'One hot'
    elif df[column].dtype=='int64':
        if unique_values < 20:
            sns.countplot(data=df, x=column)
        else:
            sns.distplot(df[column])
        proposed_encoding = 'As is'
    elif df[column].dtype=='float64':
        sns.distplot(df[column])
        if df[column].min() >= -1 and df[column].max() <=1:
            proposed_encoding = 'As is'
        else:
            proposed_encoding = 'Binning'
    else:
        cols_not_dealt_with+=1
    print(proposed_encoding)
    plt.show()
    
    ################################################################
    ## encode the variables into target dataframe ##################
    ################################################################    
    
    if proposed_encoding == 'As is':
        target_df[column]=df[column]
    
    
print('COLUMNS NOT DEALT WITH ' + str(cols_not_dealt_with))
    # distribution
    # null values
    # distinct
    # new distribution if binned
```

## Null Value Analysis

```python
df_null_values = round(df.isnull().mean()*100,2)
df_only_null_values = df_null_values.loc[(df_null_values!=0)].to_frame().reset_index()
df_only_null_values.rename(columns={'index':'row', 0:'value'}, inplace=True)
df_only_null_values.sort_values(by='value',ascending=False)
```

```python
df_only_null_values.shape
```

```python
if df_only_null_values.size > 0:
    plot_order = df_only_null_values.sort_values(by='value', ascending=False).row.values
    sns.barplot(x='row', y='value', data=df_only_null_values, order=plot_order)
    plt.xticks(rotation=90)
    plt.show()
```
